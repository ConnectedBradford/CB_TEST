```r
library(bigrquery)
bq_auth()
# Store the project ID
project_id = "yhcr-prd-bradfor-bia-core"
 
# Amend these to "point" at the datasets you want to query
 
targetdb1 <-'yhcr-prd-bradfor-bia-core.Cb_2378'
targetdb1 <-gsub(' ','',targetdb1)
print (targetdb1)
 
#can you run the above script
 
sql3 <-paste('select  *
from ',targetdb1,'.person
LIMIT 100;', sep = "")
#This runs it
tb3 <- bq_project_query(project_id, sql3)
 
#This loads  it into an R data frame
data <- bq_table_download(tb3)
data
#This displays it
#names(data)
```
